###  4. Implement the solution and test it using example inputs. Fix bugs, if any.

```python
digits = dict()
digits['2'] = ['a', 'b', 'c']
digits['3'] = ['d', 'e', 'f']
digits['4'] = ['g', 'h', 'i']
digits['5'] = ['j', 'k', 'l']
digits['6'] = ['m', 'n', 'o']
digits['7'] = ['p', 'q', 'r', 's']
digits['8'] = ['t', 'u', 'v']
digits['9'] = ['w', 'x', 'y', 'z']
```

```python
%%time
def recurs(st):
    
    result = []
    
    if len(st) == 0:
        return []
    if len(st) == 1:
        return digits[st[0]]
    if len(st) == 2:
        for x in digits[st[0]]:
            for y in digits[st[1]]:
                result.append(x+y)
    if len(st) == 3:
        for x in digits[st[0]]:
            for y in digits[st[1]]:
                for g in digits[st[2]]:
                    result.append(x+y+g)
    if len(st) == 4:
        for x in digits[st[0]]:
            for y in digits[st[1]]:
                for g in digits[st[2]]:
                    for w in digits[st[3]]:
                        result.append(x+y+g+w)
        
    return result
```

Come with the optimized correct solution and explain it in simple words below:

1. Create a dictionary containing letters for respective digits
2. If string has a single element, return a list of corresponding letters
3. If not, iterate over a digits string
4. Iterate over a list of letters for a respective number
5. Invoke function recursively to calculate the possible combinations of letters

### 8. Implement the solution and test it using example inputs. Fix bugs, if any.

```python
%%time
def recur(st):
    if len(st) == 1:
        return digits[st[0]]
    elif len(st) == 0:
        return []
    else:
        result = []
        for i in st:
            for j in digits[i]:   
                for z in recur(st[1:]):
                    result.append(j + z)

            return result
```
